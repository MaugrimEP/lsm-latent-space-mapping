```python
import numpy as np
import matplotlib.pyplot as plt

from datasets.Sarcopenie.Sarcopenie_dataset import SarcopeniaDataset
```

```python
dataset_path = "C:/Users/tmayet/Documents/datasets/fusion_preprocessed128/"
image_shape = [128, 128]
```

```python
dataset = SarcopeniaDataset(dataset_path, modes={'NONE'}, return_mode=False)

values = dict()
for i in range(len(dataset)):
    x, _ = dataset[i]
    x: np.ndarray = x.flatten()
    for pixel in x.tolist():
        if pixel in values.keys():
            values[pixel] += 1
        else:
            values[pixel] = 1
print(values)
```

```python
plt.hist(values, bins=400)
plt.show()
```
